__Open Addressing__
----------------------------------
In open addressing we have three types:
1. Linear Probing
2. Quadractic Probing
3. Double Hashing

1.LINEAR PROBING
---------------------
***
first we will make the number of slots we required and apply a Hasing function<br>
$\|\_ \|\_ \|\_ \|\_ \|\_ \|\_ \|\_ \|$ .. n slots

For example we have hasing function as $ i= key \% 7$

so initially we will make a hashsing table of 7 slots.
and then, we insert the following keys: 50,51,49,16,56,15,19

<br>

So if a location is already taken then it will start to go throught the location down until it finds a __deleted__ or __empty__ location.
Disadvantage:-- Forms Primary Clusters

-------------

3.Double Hashing
-------------
***
Double Hashing we put two hashing funcitions at once i.e;<br>
$h(key,i)= (h_{1}(key)+ih_{2}(key))\%m$

we can take hash function for the second one like,
$h_{2}(key)= 6 - key\% 6$ <br>
